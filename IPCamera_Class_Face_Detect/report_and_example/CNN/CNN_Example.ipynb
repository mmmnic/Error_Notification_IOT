```python
import dlib
import cv2
from matplotlib import pyplot as plt
import time

dlib.DLIB_USE_CUDA
```

```python
image = cv2.imread('barca_face.jpg')
hog_face_detector = dlib.get_frontal_face_detector()
cnn_face_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
```

```python
start = time.time()
## (image, number_of_windows)
faces_hog = hog_face_detector(image, 2)
end = time.time()
print("Hog + SVM Execution time: " + str(end-start))
for face in faces_hog:
    x = face.left()
    y = face.top()
    w = face.right() - x
    h = face.bottom() - y

    cv2.rectangle(image, (x,y), (x+w,y+h), (0,255,0), 2)

plt.imshow(image)
```

```python
start = time.time()
## (image, number_of_windows)
faces_cnn = cnn_face_detector(image, 2)
end = time.time()
print("CNN Execution time: " + str(end-start))
for face in faces_cnn:
    x = face.rect.left()
    y = face.rect.top()
    w = face.rect.right() - x
    h = face.rect.bottom() - y

    cv2.rectangle(image, (x,y), (x+w,y+h), (0,0,255), 2)

plt.imshow(image)
```
