```python
import cv2
import os
import numpy as np
import glob
```

```python
root_dir = 'plate_digit_text/'
all_img_paths = glob.glob(os.path.join(root_dir, '*/*.png'))
for img_path in all_img_paths:
    i = 0
    im = cv2.imread(img_path)
    # Convert to grayscale and apply Gaussian filtering
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray, (5, 5), 0)

    # Threshold the image
    ret, im_th = cv2.threshold(im_gray, 160, 255, cv2.THRESH_BINARY_INV)

#     plt.imshow(im_th)

    # Find contours in the image
    ima, ctrs, hier = cv2.findContours(im_th.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    if len(ctrs) > 0:
        for contour in ctrs:
            if ((cv2.contourArea(contour) > 20) and (cv2.contourArea(contour) < 2500)):
                # Get rectangles contains each contour
                rect = cv2.boundingRect(contour)
                x, y, width, height = cv2.boundingRect(contour)
                area = cv2.contourArea(contour)
                print(area)
                # Draw the rectangles
#                 cv2.rectangle(im, (rect[0], rect[1]), (rect[0] + rect[2], rect[1] + rect[3]), (0, 255, 0), 3) 
#                 cv2.drawContours(im, contour, 0, (0,255,0), 3)
    #             Make the rectangular region around the digit
                leng = int(rect[3] * 1.6)
                pt1 = int(rect[1] + rect[3] // 2 - leng // 2)
                pt2 = int(rect[0] + rect[2] // 2 - leng // 2)
                roi_origin = im[y:y+height, x:x+width]
#                 print(contour)
#                 cv2.imshow('origin', roi_origin)
#                 cv2.waitKey()
                print(img_path[0:len(img_path) - 4] + '_' + str(i) + '_bis.PNG')
                i = i + 1
                cv2.imwrite(img_path[0:len(img_path) - 4] + '_' + str(i) + '_bis.PNG', roi_origin)

print('done')
```
